```python
# code from class
# reading in data
in_file = "/home/eeb177-student/Desktop/eeb177/class-assignments/canids/canids.csv"
with open(in_file) as f:
    canid_records = f.readlines()[1:]
    
# use defaultdict
from collections import defaultdict

species_ranges = defaultdict(list)

# for loop to read every line
for line in canid_records: # each row is a list 
    species = line.split('","')[5] # 1a. Removing quotes from species names
    minage = line.split('","')[10]
    maxage = line.split('","')[11]
    mean_age = (float(minage) + float(maxage))/2
    #print(mean_age)
    
    # populate dictionary
    species_ranges[species].append(mean_age) #adding species

# creating empty file 
output = open("/home/eeb177-student/Desktop/eeb177/homework/canid-ranges.csv", "w")
for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    
#     print(genus, key, minage, maxage)
    outline = "{},{},{},{}\n".format(genus, key, minage, maxage)
#     output.write(genus + "," + key + "," + str(minage) + "," + str(maxage) + "\n")
    output.write(outline)
```

```python
# 3. Write a function that takes file name as argument, opens the file, 
# extracts species name with a latin binomial, ignores records that only report genus,
# populates a dictionary with species as keys and ages of fossils as values
# and returns the dictionary

from collections import defaultdict # importing modules
import re

def file_reader(file_name):
    with open(file_name, "r") as ff:
        species_ranges = defaultdict(list)
        species_records = ff.readlines()[1:]
        for line in species_records: # each row is a list 
            species = line.split('","')[5]
            if re.match("(\w+\s\w+)", species):
                species1 = line.split('","')[5]
            minage = line.split('","')[10]
            maxage = line.split('","')[11]
            mean_age = (float(minage) + float(maxage))/2
            species_ranges[species1].append(mean_age)
        return(species_ranges)
```

```python
def dict_output(my_dict,output_file):
    output = open(output_file, "w")
    for key in my_dict.keys():
        ages = my_dict[key]
        minage = min(ages)
        maxage = max(ages)
        genus = key.split(" ")[0]
    outline = "{},{},{},{}\n".format(genus, key, minage, maxage)
    output.write(outline)
```
